```python
from tika import parser
```

```python
raw = parser.from_file('HGB.pdf')
print(raw['content'])
content = raw['content']
```

```python
import re
```

```python
re.findall(r'-\sSeite\s[0-9]+\svon\s[0-9]+\s-', content)
lawtext=re.sub(r'-\sSeite\s[0-9]+\svon\s[0-9]+\s-', '', content)
```

```python
lawtext=re.sub(r'Ein\sService\sdes\sBundesministeriums\sder\sJustiz\sund\sfür\sVerbraucherschutz', '', lawtext)
lawtext=re.sub(r'sowie\sdes\sBundesamts\sfür\sJustiz\s‒\swww.gesetze-im-internet.de', '', lawtext)
print(lawtext)
```

```python
with open('HGB.txt', 'w') as f:
    print(lawtext, file=f) 
```

```python
import nltk
wsj = sorted(set(nltk.corpus.treebank.words()))
fd = nltk.FreqDist(vs for word in wsj
    for vs in re.findall(r'[aeiou]{2,}', word))

fd.items()
```

```python
lawtext=lawtext.decode('utf-8')
tokens = nltk.word_tokenize(lawtext)
tokens
```

```python
fdist = nltk.FreqDist(tokens)
for key in fdist:
    print(key + ':', fdist[key], end='; ')
```
